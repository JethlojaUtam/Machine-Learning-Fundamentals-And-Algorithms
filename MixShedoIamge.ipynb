```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('Ulogo.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Creating the kernel
kernel1=np.array([[1,0,0],[0,0,0],[0,0,-1]])
kernel2=np.array([[-1,-1,0],[-1,0,1],[0,1,1]])
kernel3=np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
kernel4=np.array([[0,0,1],[0,0,0],[-1,0,0]])

#Applying the kernel
img1=cv2.filter2D(img,-1,kernel1)
img2=cv2.filter2D(img,-1,kernel2)
img3=cv2.filter2D(img,-1,kernel3)
img4=cv2.filter2D(img,-1,kernel4)

#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(2,3,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(2,3,2),plt.imshow(img1),plt.title('Shedo-1 Image'),plt.axis('off')
plt.subplot(2,3,3),plt.imshow(img2),plt.title('Shedo-2 Image'),plt.axis('off')
plt.subplot(2,3,4),plt.imshow(img3),plt.title('Shedo-3 Image'),plt.axis('off')
plt.subplot(2,3,5),plt.imshow(img4),plt.title('Shedo-4 Image'),plt.axis('off')
```
