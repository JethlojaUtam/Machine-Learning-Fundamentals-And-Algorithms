```python
#Mix of MotionBlur Function Image

import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#(i) Horizontalblur function
size=30
kernel1=np.zeros((size,size))
kernel1[int(size/2),:]=np.ones(size)
kernel1=kernel1/size

#(ii) Verticalblur function
size=30
kernel2=np.zeros((size,size))
kernel2[:,int(size/2)]=np.ones(size)
kernel2=kernel2/size

#Applying the kernel
horizontal=cv2.filter2D(img,-1,kernel1)
vertical=cv2.filter2D(img,-1,kernel2)

#Ploting the image
plt.subplot(1,3,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(horizontal),plt.title('Horizontalblur Image'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(vertical),plt.title('Verticalblur Image'),plt.axis('off')
```
