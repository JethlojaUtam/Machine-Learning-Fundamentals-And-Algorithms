```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Median blur function
median=cv2.medianBlur(img,(19)) #It can be any odd number

#Ploting the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Iamge'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(median),plt.title('Medianblur Image'),plt.axis('off')
```
