```python
#Horizontal of MotionBlur Function Image

import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#(i) Horizontalblur function
size=30
kernel=np.zeros((size,size))
kernel[int(size/2),:]=np.ones(size)
kernel=kernel/size

#Applying the kernel
horizontal=cv2.filter2D(img,-1,kernel)

#Ploting the image
plt.subplot(2,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(horizontal),plt.title('Horizontalblur Image'),plt.axis('off')
```

```python
#Vertical of MotionBlur Function Image

import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#(ii) Verticalblur function
size=30
kernel=np.zeros((size,size))
kernel[:,int(size/2)]=np.ones(size)
kernel=kernel/size

#Applying the kernel
vertical=cv2.filter2D(img,-1,kernel)

#Ploting the image
plt.subplot(2,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(vertical),plt.title('Verticalblur Image'),plt.axis('off')
```
