```python
import cv2
import matplotlib.pyplot as plt

#Loading an image
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intensity value, not on color, so BGR2GRAY is used

#Bluring on image using boxblur function
box=cv2.blur(img,(44,44))

#Canny(src,threshold1,threshold2)
canny=cv2.Canny(img,threshold1=20,threshold2=40)
canny_after_blur=cv2.Canny(box,threshold1=20,threshold2=40)

#Ploting the image
plt.subplot(1,3,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(canny,cmap='gray'),plt.title('Canny Image'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(canny_after_blur),plt.title('Canny-After-Blur Image'),plt.axis('off')
```
