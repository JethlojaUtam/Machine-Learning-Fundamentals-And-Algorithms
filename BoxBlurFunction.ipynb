```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Box blur function
box=cv2.blur(img,(20,20)) #We can have positive number

#Ploting the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(box),plt.title('Boxblur Image'),plt.axis('off')
```
