```python
import cv2
import matplotlib.pyplot as plt

#Loading the image
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intersity value, not on color, so BGR2GRAY is used

#Sobel(src,ddepth,dx,dy,kernel_size)
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,10)

#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(sobelx,cmap='gray'),plt.title('Soble-x Image'),plt.axis('off')
```

```python
#Combine Image of sobel-x and sobel-y

import cv2
import matplotlib.pyplot as plt

#Loading the image
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intersity value, not on color, so BGR2GRAY is used

#Sobel(src,ddepth,dx,dy,kernel_size)
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,10)
sobel_combine=cv2.magnitude(sobelx,sobely) #magnitude combines stong edges from sobel-x and sobel-y

#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(2,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(sobelx,cmap='gray'),plt.title('Soble-x Image'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(sobely,cmap='gray'),plt.title('Sobel-y Image'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(sobel_combine,cmap='gray'),plt.title('Sobel-Combine Image'),plt.axis('off')
```
