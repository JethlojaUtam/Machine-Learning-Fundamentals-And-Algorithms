```python
# Erosion and Dilation Image

import cv2
import matplotlib.pyplot as plt
import numpy as np

#Load image
img=cv2.imread('one.jpg')

#Creating the kernel
kernel=np.ones((3,3),np.uint8)

#Appling the Dilation
dilation=cv2.dilate(img,kernel,iterations=1)

#Ploting the image
plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(dilation,cmap='gray'),plt.title('Diletion Image'),plt.axis('off')
```
