```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Creating the kernel
kernel=np.array([[-1,-1,-1],[-1,10,-1],[-1,-1,-1]])

#Applying the kernel
img1=cv2.filter2D(img,-1,kernel)

#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(img1),plt.title('Sharpening Image'),plt.axis('off')
```
