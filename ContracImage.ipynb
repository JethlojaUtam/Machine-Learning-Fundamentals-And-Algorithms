```python
#Contarc base image
#Contracing the Gray scale image
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('Utam.jpg',0)
img1=cv2.equalizeHist(img)

#Ploting the image
plt.subplot(1,2,1),plt.imshow(img,cmap='gray'),plt.title('Sourece Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(img1,cmap='gray'),plt.title('EqulizeHist Image'),plt.axis('off')
```
