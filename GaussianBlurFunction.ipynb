```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Gaussian blur function
gaussian=cv2.GaussianBlur(img,(35,35),9) #It can be any odd number, 0 mean automatically calculate the blur effect(you can have number instead of 0)

#Ploting the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(gaussian),plt.title('Gaussianblur Image'),plt.axis('off')
```
