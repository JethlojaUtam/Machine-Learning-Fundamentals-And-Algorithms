```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#Box blur function
box=cv2.blur(img,(20,20)) #We can have any positive number
             
#Gaussian blur function
gaussian=cv2.GaussianBlur(img,(35,35),9) #It can be any odd number, 0 mean automatically calculate the blur effect(you can have number instead of 0)

#Median blur function
median=cv2.medianBlur(img,(19)) #It can be any odd number

#Ploting the image
plt.subplot(2,2,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(2,2,2),plt.imshow(box),plt.title('Boxblur Image'),plt.axis('off')
plt.subplot(2,2,3),plt.imshow(gaussian),plt.title('Gaussianblur Image'),plt.axis('off')
plt.subplot(2,2,4),plt.imshow(median),plt.title('Medianblur Image'),plt.axis('off')
```
