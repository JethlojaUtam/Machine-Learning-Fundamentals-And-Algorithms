```python
import cv2
import matplotlib.pyplot as plt

#Loading the image
img=cv2.imread('Utam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intersity value, not on color, so BGR2GRAY is used

#Bluring an image using gaussianblur function
gaussian=cv2.GaussianBlur(img,(3,3,),0)
laplacian_after_blur=cv2.Laplacian(gaussian,cv2.CV_64F,ksize=3)

#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,3,1),plt.imshow(img),plt.title('Source Image'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(gaussian,cmap='gray'),plt.title('Gaussianblur Image'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(laplacian_after_blur,cmap='gray'),plt.title('Laplacian Image'),plt.axis('off')
```
